```python
category_names = sorted(category_names)
category_dict = {key.lower():[] for key in category_names}
category_dict
```

```python

for category in category_dict:
    print(category)
    
    if '/' in category:
        cat = category.split('/')
        
        for idx in range(df.shape[0]):
            if cat[0] in df.Comment[idx] or cat[1] in df.Comment[idx]:
                category_dict[category].append(idx)     
                
    else:
        for idx in range(df.shape[0]):

            if category in df.Comment[idx]:
                category_dict[category].append(idx)

print('donee')
```

```python
# Create a Pandas Excel writer using XlsxWriter as the engine.o
writer = pd.ExcelWriter('reviews_each_category.xlsx', engine='xlsxwriter')

for category in category_dict:
    print(category)
    category_df = pd.DataFrame({'Comment':df.loc[category_dict[category], 'Comment']})
    
    sheet_name = category
    if '/' in  sheet_name:
        sheet_name = sheet_name.split('/')[0]
    category_df.to_excel(writer, sheet_name=sheet_name, index=False)
    
# Close the Pandas Excel writer and output the Excel file.
writer.save()
print('\nDONEE')
```

### treat 'hati-hati'

```python
hati2 = ['hati"', 'hati-hati', 'hati2']
for i in range(len(comment)):
    for word in hati2:
        comment[i] = comment[i].replace(word, 'hatihati')
print(len(comment))
```

```python
words = []
for i in range(len(comment)):
    sentence = re.sub(r'[,./]+', ' ', comment[i])
    sentence = re.sub(r'\s+', ' ', sentence)
    word = sentence.split()
    words.extend(word)

print(len(words))
words = list(set(words))
print(len(words))
```

### cleansing

```python
ONLY_ALPHABET = re.compile('[^a-z]')

# only keep an alphabet
comment = [ONLY_ALPHABET.sub(' ', sent) for sent in comment]
# remove long space
comment = [re.sub('\s{2,}', ' ', sent) for sent in comment]
# remove empty sentence
comment = [sent for sent in comment if sent.strip()]
# remove duplicate sentence
comment = list(set(comment))

print(len(comment))
```
